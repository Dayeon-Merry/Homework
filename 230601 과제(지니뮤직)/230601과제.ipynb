```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 지니 차트 200위까지 크롤링
# 반복문을 사용하여 크롤링(페이지 번호)
# 수집한 데이터를 데이터프레임으로 만들고 저장
# 엑셀로 내보내기
# 과제는 github에 업로드
df = []
for i in range(1, 5):
  header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
  request = requests.get(f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=17&rtm=Y&pg={i}', headers=header)
  soup = BeautifulSoup(request.text)
  info = soup.find_all('td', {'class':'info'})
  num = soup.find_all('td', {'class':'number'})
  for z, a in zip(info, num):
    title = z.find('a', {'class':'title'}).text.strip()
    artist = z.find('a',{'class':'artist'}).text.strip()
    lank = a.text.split('\n')[0]
    dic = {'순위': lank, '제목': title, '가수': artist}
    df.append(dic)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df_fin = pd.DataFrame(df)
df_fin
```

```python
df_fin.to_excel('지니뮤직 Top200 크롤링.xlsx')
```
